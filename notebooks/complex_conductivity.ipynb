```python
import numpy as np
from scipy.constants import epsilon_0
import matplotlib.pyplot as plt
```

```python
# def cole_cole(sigma_dc, epsilon_hf, )
f = np.logspace(1, 5, 81)
omega = 2 * np.pi * f
epsilon_hf =  2
d_epsilon_iw = 500
f_iw = 1000
tau_iw = 1./(2*np.pi*f_iw)
alpha_iw = 0.1

d_epsilon_ice = 100
f_ice = 1e4
tau_ice = 1./(2*np.pi*f_ice)
alpha_ice = 0.

sigma_dc = 1e-5
epsilon_iw = (d_epsilon_iw / (1+(1j*omega*tau_iw)**(1.-alpha_iw)))
epsilon_ice = (d_epsilon_ice / (1+(1j*omega*tau_ice)**(1.-alpha_ice))) 
epsilon_sig =  - 1j*sigma_dc/(omega * epsilon_0)
epsilon = epsilon_hf  + epsilon_ice + epsilon_iw + epsilon_sig
# epsilon = epsilon_ice + epsilon_iw + epsilon_sig

sigma_eff = epsilon * 1j  * omega * epsilon_0
sigma_ice = epsilon_ice * 1j  * omega * epsilon_0
sigma_iw = epsilon_iw * 1j  * omega * epsilon_0

rho_eff = 1./sigma_eff
rho_iw = 1./sigma_iw
rho_ice = 1./sigma_ice
```

```python
plt.loglog(f, -epsilon.imag, label='total')
plt.loglog(f, -epsilon_iw.imag, label = 'iw')
plt.loglog(f, -epsilon_ice.imag, label = 'ice')
plt.loglog(f, -epsilon_sig.imag, label = 'sig')
plt.legend()
```

```python
plt.semilogx(f, np.rad2deg(np.arctan(rho_eff.imag/rho_eff.real)))
```
